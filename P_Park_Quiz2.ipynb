##Exploratory Data Analysis##
Synopsis of Automobile Attributes:
1. Symboling- A ranking system that gauges the risk as it relates to the price indicated. The range is from -3 to 3. -3 is the safest, while +3 is the riskiest

2. Normalized_losses- This is the relative average loss payment per the insured vehicle year. The value is normalized in reference to all other automobiles. The range is from 65 - 256.

3. Make- This is the maker of the brand. (alfa_romero, audi, bmw, chevrolet, homda, isuzu, dodge, jaguar, mazda, mercedes-benz, mercury, mitsubishi, nissan, peugot, plymouth, porsche, renault, saab, subaru, toyota, volkswagen, volvo)

4. Fuel_type- The type of fuel that the car uses. (diesel, gas)

5. Aspiration- std, turbo.

6. Number of doors- Numbers of doors. (four, two)

7. Body_style- The body type. (hardtop, wagon, sedan, hatchback, convertible)

8. Drive_wheels- The numbers of wheels for each car. (4wd, fwd, rwd)

9. Engine_location- Location of the engine. (Front, rear)

10. Wheel_base- The distance between the centers of the front and rear wheels. (86.6 - 120.9)

11. Length- Length of the car. (141.1 - 208.1)

12. Width- The width of the car. (60.3 - 72.3)

13. Height- Height of the car. (47.8 to 59.8)

14. Curb-weight- The weight of a car without passenegers or baggage. (1488 to 4066)

15. Engine_type- Type of engine. (dohc, dohcv, I, ohc, ohcf, ohcv, rotor)

16. Num_of_xcylinders- Numbers of cylinders. (eight, five, four, six, three, twelve, two)

17. Engine_size- Size of engine. (61 - 326)

18. Fuel_system- Fuel systems that cars use. (1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi)

19. Bore- Size of the cylinder for each car. (2.54 - 3.94)

20. Stroke- The distance travelled by the piston in each cycle (2.07 - 4.17)

21. Compression-ratio- The ratio of the max and min volume in the cylinder of an internal combustion engine. (7 - 23)

22. Horsepower- Engine's power. (48 - 288)

23. Peak_rpm- The maximum revolutions per Minute. (4150 to 6600)

24. City_mpg- This is our target variable. It is the average MPG for a car as it travels in the city. (13 - 49)

25. Highway_mpg- The average MPG for a car on the highway. (16 - 54)

26. Price- Price of the car. (5118 - 45400)

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

checkNulls = df.isnull().sum()

for col in df.columns:
    
    df[col].replace({'?':np.nan},inplace=True)
    

df['wheelbase'] = df['wheelbase'].fillna(0)
df['length'] = df['length'].fillna(0)
df['width'] = df['width'].fillna(0)
df['height'] = df['height'].fillna(0)
df['curbheight'] = df['curbheight'].fillna(0)
df['enginesize'] = df['enginesize'].fillna(0)
df['compressionratio'] = df['compressionratio'].fillna(0)
df['horsepower'] = df['horsepower'].fillna(0)
df['peakrpm'] = df['peakrpm'].fillna(0)
df['citympg'] = df['citympg'].fillna(0)


df.isnull().sum()
newDf = df[['wheelbase', 'length', 'width', 'height', 'curbheight', 'enginesize', 'compressionratio', 'horsepower', 'peakrpm', 'citympg']].copy()
pca_test = newDf.drop('wheelbase', 1)

data_scaled = pd.DataFrame(preprocessing.scale(pca_test), columns = pca_test.columns)

pca = PCA(n_components = len(data_scaled.columns))
x_new = pca.fit_transform(data_scaled)
x_new = pd.DataFrame(x_new)
x_new
```
